```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

img_path = "../images/huang1.jpg"

img = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
```

```python
def filtro_suav(f,x,y):
    return 1/2 * (f[y,x] + 1/4 * (f[y,x-1] + f[y,x+1] + f[y-1,x] + f[y+1,x]))


img_suav = img.copy()

print(img.shape)

for y in range(1,img_suav.shape[1]-1):
    for x in range(1,img_suav.shape[0]-1):
        img_suav[y,x] = filtro_suav(img,x,y)


plt.figure()
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")

plt.subplot(1,2,2)
plt.imshow(img_suav,cmap="gray")
```
